```python
files = sorted(glob.glob('traj*.pkl'))
```

```python
def get_water_atom_index(traj):
    topology = traj.topology
    water_atom_index = []
    for atom in topology.atoms:
        if (atom.element.symbol =='O'):
            water_atom_index.append(atom.index)

    water_atom_index = np.array(water_atom_index)
    return water_atom_index

def get_distance(water_atom_index,traj):
    atom_pair = np.vstack((water_atom_index,np.ones_like(water_atom_index)*8570)).T
    distance= md.compute_distances(traj,atom_pairs=atom_pair)
    dist = np.sort(distance)
    dist = dist[:,:100]

    sorted_water_index = np.empty_like(dist)
    for j in range(len(dist[:,0])):
        for i in range(len(dist[0])):
            sorted_water_index[j,i]=np.where(dist[j,i]==distance[j,:])[0]

    return sorted_water_index

def get_xyz_water(traj,sorted_water_index):
    xyz_sorted_water=traj.xyz[tuple(sorted_water_index.astype(int))][:]
    return xyz_sorted_water

# def get_fullenere_center():
#     fullerene_atom_index = np.load('fullerene_atom_index.npy')
#     traj.xyz[]
```

```python
for f in files:
    trajs = pickle.load(open(f,'rb'))
    #cavity_xyz = []; ligand_xyz = [];
    xyz_water_sorted = []
    # xyz_clw_sorted.append(cavity_xyz)
    # xyz_clw_sorted.append(ligand_xyz)
    for i in range(30):
        water_atom_index = get_water_atom_index(trajs[i])  ## define the function
        sorted_water_index = get_distance(water_atom_index,trajs[i])
        temp_xyz_water = get_xyz_water(trajs[i],sorted_water_index)
        xyz_water_sorted.append(temp_xyz_water)
        del temp_xyz_water, dist
        gc.collect()
    pickle.dump(xyz_water_sorted,open('xyz_water_sorted.pkl','wb'))
    del xyz_water_sorted
    gc.collect()

    
    
```
